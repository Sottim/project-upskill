```python
import tiktoken
from transformers import AutoTokenizer
```

```python
encoding = tiktoken.get_encoding("cl100k_base")
```

```python
tokenizer = AutoTokenizer.from_pretrained("gpt2")
```

```python
sentences = [
    "Hello, world!",
    "I love building LLMs.",
    "ChatGPT is amazing.",
    "The quick brown fox jumps over the lazy dog.",
    "Transformers changed AI forever!"
]

for sentence in sentences:
    # Tokenize using tiktoken
    tiktoken_tokens = encoding.encode(sentence)
    print(f"Sentence: {sentence}")
    print(f"tiktoken tokens: {tiktoken_tokens}")
    # Print number of words in the sentence
    num_words = len(sentence.split())
    print(f"Number of words: {num_words}")
    # Print number of tokens in the sentence
    num_tokens = len(tiktoken_tokens)
    print(f"Number of tokens from tiktoken: {num_tokens}")

    # Tokenize using Hugging Face Transformers
    hf_tokens = tokenizer.encode(sentence, add_special_tokens=False)
    print(f"Hugging Face tokens: {hf_tokens}")

    # Print number of tokens in the sentence
    num_tokens_hf = len(hf_tokens)
    print(f"Number of tokens from Hugging Face: {num_tokens_hf}")
    print("-" * 50)
```

```python
sentence = "Transformers changed AI forever!"

# OpenAI
openai_ids = encoding.encode(sentence)
openai_tokens = [encoding.decode([i]) for i in openai_ids]

# Hugging Face
hf_ids = tokenizer.encode(sentence)
hf_tokens = tokenizer.convert_ids_to_tokens(hf_ids)

print("=" * 70)
print(sentence)

print("\nOpenAI")
print(openai_ids)
print(openai_tokens)

print("\nGPT-2")
print(hf_ids)
print(hf_tokens)
```
